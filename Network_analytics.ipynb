# Network Analytics

```python
import pandas as pd
import networkx as nx
```

```python
G = pd.read_csv("D:/Data Scientist/Network Analytics/Datasets/routes.csv")
```

```python
G
```

```python
G = G.iloc[0:100, 1:10]
G
```

```python
G.info()
```

```python
g = nx.Graph()
g
```

```python
g = nx.from_pandas_edgelist(G, source = 'Source Airport', target = 'Destination Airport')
```

```python
print(g)
```

```python
# Degree Centrality
d = nx.degree_centrality(g)
print(d)
```

```python
pos = nx.spring_layout(g)
nx.draw_networkx(g, pos, node_size = 15, node_color = 'red')
```

```python
# closeness centrality
closeness = nx.closeness_centrality(g)
print(closeness)
```

```python
## Betweeness Centrality 
b = nx.betweenness_centrality(g) # Betweeness_Centrality
print(b)
```
